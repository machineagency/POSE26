## Taking Pictures with the Camera tool

### Before Starting:
- Clear any existing items off the bed of your machine!

```python
from science_jubilee.Machine import Machine
from science_jubilee.tools.Camera import Camera
```

```python
m = Machine(address='10.18.178.192')
```

```python
# m.home_all() # uncomment and run if your machine needs to be homed
```

### Tool import and set up

Our camera tool will require a `tool_index` and a `tool_name`. These should be the same as the ones defined in the machine's `config.g` file. 

**Camera Tool**

```python
# Initialize the camera from a configuration file
cam = Camera.from_config(1, 'Camera', 'camera_opencv.json')
# note that 1 is the tool index, which is different than camera_index in the config file
# camera_index is useful when you have multiple USB cameras

# Load the camera tool into the machine
m.load_tool(cam)
```

Let's pick it up!

```python
m.pickup_tool(cam)
```

Let's move the camera to the spot where you want a picture taken.

```python
m.move_to(x=150, y=150, z=100)  # move to the center
```

```python
cam.show_image()  # display the camera view inline
```

Let's take a photo!

```python
filename = 'test_image.jpg'
cam.capture(filename)
```

Simple as is!

Let's take a video!

```python
cam.record('test_video.mp4', 5)  # record a 5 second video
```

Sometimes it might be helpful to record videos in parallel of other things! you can use `record_video_async` for that purpose.

```python
cam.start_recording('test_video_async.mp4', 5)
# record a 5 second video asynchronously
# while the bed is moving
m.move(dx=30, dy=30)
m.move(dx=-30, dy=-30, dz=30)
# note that if you move z at full speed, the frame may be out of focus
```

We can now park the camera:

```python
m.park_tool()
```

And turn off the camera!

```python
cam.release()
```
